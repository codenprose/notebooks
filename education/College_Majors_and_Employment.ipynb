```python
# Create function to calculate the total number of students in each major category

def calculate_major_categories(df):
    # Create list of all the major categories
    categories = df["Major_category"].value_counts().index
    # Create dictionary to store major category totals
    major_category_totals = dict()
    
    # Loop through each major category
    for category in categories:
        # Search the Major_catgory column in dataframe for matches to current category
        major = df[df["Major_category"] == category]
        # Calculate toal number of students in current major.
        total = major["Total"].sum()
        # Add total to major category totals dictionary
        major_category_totals[category] = total
        
    return major_category_totals

# Get totals for all_ages dataset
calculate_major_categories(all_ages)
```

```python
# Get totals for recent_grads dataset
calculate_major_categories(recent_grads)
```
